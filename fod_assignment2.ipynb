```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
A=np.array([
    [2,-1],
    [0,3]
])
B=np.array([
    [1,2],
    [-3,4]
])
alpha=3
beta=2
print(A+B)
print(A*B)
print(alpha*A-beta*B)
```

```python
#determinant
det_val = np.linalg.det(A)
print("Determinant:", det_val)
```

```python
#inverse of matrix
A_inv = np.linalg.inv(A)

print("Inverse of A:")
print(A_inv)
```
