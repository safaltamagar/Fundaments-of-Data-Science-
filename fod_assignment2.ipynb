```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
A=np.array([
    [2,-1],
    [0,3]
])
B=np.array([
    [1,2],
    [-3,4]
])
alpha=3
beta=2
print("A+B:")
print(A+B)
print("A*B:")
print(A*B)
print("3A-2B:")
print(alpha*A-beta*B)
```

```python
#determinant
det_val = np.linalg.det(A)
print("Determinant:", det_val)
```

```python
#inverse of matrix
A_inv = np.linalg.inv(A)

print("Inverse of A:")
print(A_inv)
```

```python
#eigenvalues and eigenvectors
C=np.array([
    [4,1],
    [-2,1]
])

eigenvalues, eigenvectors = np.linalg.eig(C)

print("Matrix C:")
print(C)
print("Eigenvalue (first):")
print(eigenvalues[0])
print("Corresponding Eigenvector:")
print(eigenvectors[:, 0])
```

```python
# Feature matrix
X=np.array([
    [4,1.5,100], #car 1
    [6,2.0,150]  #car 2
])

# Price vector
y=np.array([20,30])

#X.T*B=y
#B=(X.T*X)^-1 X^T y
#X.T is transpose matrix
beta=np.linalg.pinv(X.T @ X) @ X.T @ y #coefficients

# Features of car 3
car3=np.array([8, 3.0, 200])

# Predicted price of car 3
price_car3=car3 @ beta
print("Predicted price:",price_car3)
```
