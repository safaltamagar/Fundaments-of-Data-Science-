Data Visualization Using Health Dataset(Health_heart_experimental)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Data:

```python
df=pd.read_csv("Health_heart_experimental.csv")
df.head(10)
```

The above table is an example of how the dataset is written.

lineplot:

```python
sns.lineplot(x='BMI',y='DiaBP',data=df)
```

Normally,as BMI increases, diastolic pressure increases.In our survey,we
can see that there are more people who have low BMI and high Diastolic BP.
People with low bmi and high diabp have hypertension.High BMI together
with high diastolic blood pressure (DiaBP) indicates an increased 
cardiovascular and metabolic health risk and heart attack.

Bar Plot:

```python
sns.barplot(x='sex',y='HR',data=df)
```

The above bar plot shows that the average heart rate is same for both
gender.The above bar plot suggests that the gender of a person does not indicate the heart rate of a person. However, there may be slight differences that the bar plot has failed to take into account since we are looking at the average  heart rate.

Pie Chart:

```python
plt.title('Health Indication Distribution')
df['indication'].value_counts().plot(kind='pie', autopct='%1.2f%%')
```

```python
In above piechart,0 indicates healthy BMI range and 1 indicates unhealthy 
BMI rage.Indication is a categorical label that indicates a personhealth 
condition based on BMI and diastolic blood pressure.The pie chart 
reveals that individuals with high BMI frequently also exhibit high 
diastolic blood pressure, indicating a strong association
with increased cardiovascular risk.
```

Boxplot:

```python
sns.boxplot(df['DiaBP'])
```

In the above box plot, we can see that the median value for diastolic bp
lies between 100 and 90.In our dataset, we have people that have DiaBP 
ranging from 70 to 120.From our dataset, we can see that people have
high diaBP which is health risk as it increases the risk of heart attack. 

Histogram:

```python
plt.hist(df['SysBP'])
```

From the above histogram, we can see that the people have a balanced mix of lower and higher systolic BP values.We can assume that bp measurement
are consistebnt and stable across all people.A symmetrical histogram indicates that systolic blood pressure values are normally distributed around a central value with no significant skewness.

Scatter Plot:

```python
df = pd.read_csv("Health_heart_experimental.csv")
sns.scatterplot(x='age',y='DiaBP',hue='age',data=df)
plt.show()
```

An even distribution of points across the plot indicates that DiaBP values are spread fairly uniformly across all age groups.There is no strong positive or negative correlation between age and diastolic blood pressure.Age alone does not appear to be a strong indicator of diastolic blood pressure in this dataset.
